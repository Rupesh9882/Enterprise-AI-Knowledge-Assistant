Loading Embedding Modal ( from Hugging face)

```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np

model = SentenceTransformer("all-MiniLM-L6-v2")
```

Creating example Documents

```python
docs = [
"Artificial intelligence is transforming industries.",
"Cloud computing enables scalable applications.",
"Large language models can answer complex questions.",
"Vector databases help with semantic search."
]
```

Create Embeddings

```python
vectors = model.encode(docs)
```

Create Vector Database Using FAISS

```python
index = faiss.IndexFlatL2(len(vectors[0]))
index.add(np.array(vectors))
```

Query the System (mini semantic search engine.)

```python
query = model.encode(["What is AI?"])

D, I = index.search(np.array(query), 1)

print(docs[I[0][0]])
```
